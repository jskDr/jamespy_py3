```python
from qiskit import QuantumCircuit, Aer, assemble
import numpy as np
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.visualization import plot_state_qsphere, plot_bloch_multivector
from qiskit.visualization import array_to_latex

from IPython.display import display
import IPython
```

## Common Library

```python
Aer.backends()
```

```python
def show_statevector(qc):
    svsim = Aer.get_backend('aer_simulator')
    qc.save_statevector()
    qobj = assemble(qc)
    result = svsim.run(qobj).result()
    final_state = result.get_statevector()
    IPython.display.display(plot_state_qsphere(final_state))
    IPython.display.display(plot_bloch_multivector(final_state))
    return array_to_latex(final_state, prefix="\\text{Statevector} = ")
```

## Superposition checking

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.h(0)
show_statevector(qc)
```

```python
qc = QuantumCircuit(1)
#qc.x(0)
qc.h(0)
show_statevector(qc)
```
