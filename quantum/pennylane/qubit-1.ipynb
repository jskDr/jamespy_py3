```python
import pennylane as qml
from pennylane import numpy as np

import matplotlib.pyplot as plt
```

## Shot 1

```python
dev = qml.device("default.qubit", wires=1, shots=1)

@qml.qnode(dev, interface='autograd')
def circuit(phi):  
    qml.RX(phi, wires=0)
    return qml.expval(qml.PauliZ(0))

def cost(phi):
    return np.abs(circuit(phi) - 1.0)**2

opt = qml.GradientDescentOptimizer(stepsize=0.05)

steps = 200
phi = np.pi

phi_list = []
for i in range(steps):
    # update the circuit parameters
    #print(i, phi)
    phi = opt.step(cost, phi)
    # phi = np.mod(phi, 2*np.pi)
    phi_list.append(phi)

plt.figure(figsize=[10,20])
plt.subplot(3,1,1)
plt.plot(phi_list)
plt.title('phi')
plt.grid()
plt.subplot(3,1,2)
plt.plot([cost(phi) for phi in phi_list])
plt.title('cost(phi)')
plt.grid()
plt.subplot(3,1,3)
plt.grid()
plt.plot([circuit(phi) for phi in phi_list])
plt.title('circuit(phi)')
```
