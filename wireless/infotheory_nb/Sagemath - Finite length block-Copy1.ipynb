# Asymptotic Performance of Finte Length Block Code

```python
pp = pretty_print
```

```python
var('n P epsilon t')
pp(epsilon)
```

```python
C(P) = (1/2)*log(1+P,2)
pp(C)
# plot(C(P), title='$C(P)$', title_pos=(0.5,-0.05), axes_labels=[r'$\theta$', r'$\sin(\theta^2)$'])
plot(C(P), (P,0,10), axes_labels=[r'$P$', r'$C(P)$'])
```

```python
V(P) = (P/2) * (P+2) / (P+1)**2 * log(e,2)**2
pp(V)
```

```python
rho(n) = (1/2)*log(n,2)
pp(rho)
```

```python
q(t) = (1/sqrt(2*pi))*e**((-t**2)/2)
pp(q)
plot(q(t),(t,-3,3)).show()
Q(x) = integral(q(t),(t,x,oo))
pp(Q)
ans = solve(Q(x) == epsilon, x)
invQ(epsilon) = ans[0].rhs()
pp(invQ)
```

```python
M(n,P,epsilon) = n*C(P) - sqrt(n*V(P))*invQ(epsilon) + rho(n)
pp(M)
R(n,P,epsilon) = M(n,P,epsilon)/n
```

```python
p1 = plot(R(n,1,0.001),(n,1,2000))
p2 = line([(0,C(1)),(2000,C(1))], linestyle='--')
p = p1 + p2
p.show(ymin=0, ymax=0.55, gridlines=True)
```
