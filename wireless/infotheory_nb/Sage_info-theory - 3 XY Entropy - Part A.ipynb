```python
def entropy_x_and_y(Px_and_y_2d_list):
    h = 0
    for Px_row_list in Px_and_y_2d_list:
        for Px_and_y in Px_row_list:
            p = Px_and_y
            h += -p*log(p,2)
    return h    

h = entropy_x_and_y([[Rational('1/8'),Rational('3/4')],[Rational('1/4'),Rational('1/4')]])
pretty_print(h)
```
