## Frozen case
- Encoding시만 frozen을 적용한 경우이다. 이 경우는 약간의 성능만 좋아진다. 왜냐하면 error propagation은 포함되어 있기 때문이다. 

```python
polar = nb_polar.PolarCodeFrozen(N_code=8, K_code=4, frozen_flag_n=np.array([1,1,1,1,0,0,0,0]))
polar.run(SNRdB_list=list(range(10)), N_iter=100000, flag_fig=True)    
```

```python
polar = nb_polar.PolarCodeFrozen(N_code=8, K_code=4, frozen_flag_n=np.array([1,1,1,1,0,0,0,0]))
polar.run(SNRdB_list=list(range(10)), N_iter=100000, flag_fig=True)    
```
