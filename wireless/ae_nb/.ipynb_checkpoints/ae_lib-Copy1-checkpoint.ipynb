```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Layer, BatchNormalization
from tensorflow.keras import Sequential

import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

print(tf.__version__)

%load_ext autoreload
%autoreload

from wireless import ae
```

```python
ae.ae_train_test(
N_mod_bits=2,
Code_K=2,
Code_N=4,
N_sample=1000,
N_episodes=100,
EbNo_dB=10)   
```

```python
EbNo_dB_l = range(10)
BER_l = []
for EbNo_dB in EbNo_dB_l:
    trained_model = ae.ae_train_test(N_mod_bits=1,Code_K=2,Code_N=2,N_sample=2000,N_episodes=500,EbNo_dB=EbNo_dB)   
    BER = ae.large_test(trained_model, N_sample=100000)
    BER_l.append(BER)
    print(EbNo_dB, BER)    

plt.semilogy(EbNo_dB_l, BER_l)
plt.xlabel('SNR(dB)')
plt.ylabel('BER')
plt.title('AE transceiver: $(N,K)=(2,2)$, BPSK')
plt.grid()
plt.show()    
    
BER_l
```
